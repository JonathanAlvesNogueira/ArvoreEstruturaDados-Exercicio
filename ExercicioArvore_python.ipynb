### Desenvolva uma aplicação de recomendação de filmes que sugira pelo menos três filmes para cada faixa etária (livre, 10+, 12+, 14+, 16+, 18+). Use a abordagem MVP para lançar uma versão inicial simples e funcional, permitindo feedback para melhorias.

```python
#  Classe Vertice
class Node: 
    # Recebe no construtor a Idade, e os filhos do vertice (nó)
    def __init__(self, value): 
        self.value = value 
        self.children = [] 

    # Determina os filhos do vertice
    def add_child(self, child_node): 
        self.children.append(child_node) 

    # Retorna os filhos do vertice
    def get_children(self): 
        return self.children 
```

```python
root = Node("Raiz") 

# Cria os Objetos nó passando como valor a Faixa Etária dos filmes
livre = Node("Livre") 
dez_anos = Node("10+") 
doze_anos = Node("12+") 
quatorze_anos = Node("14+")
dezesseis_anos = Node("16+")
dezoito_anos = Node("18+")

# Adicionar nós de faixa etária à raiz 
root.add_child(livre) 
root.add_child(dez_anos) 
root.add_child(doze_anos) 
root.add_child(quatorze_anos)
root.add_child(dezesseis_anos)
root.add_child(dezoito_anos)

# Adicionar filmes a cada nó de faixa etária 
livre.add_child(Node("Filme Livre 1")) 
livre.add_child(Node("Filme Livre 2")) 
livre.add_child(Node("Filme Livre 3")) 

dez_anos.add_child(Node("Filme 10+ 1")) 
dez_anos.add_child(Node("Filme 10+ 2")) 
dez_anos.add_child(Node("Filme 10+ 3")) 

doze_anos.add_child(Node("Filme 12+ 1")) 
doze_anos.add_child(Node("Filme 12+ 2")) 
doze_anos.add_child(Node("Filme 12+ 3"))

quatorze_anos.add_child(Node("Filme 14+ 1"))
quatorze_anos.add_child(Node("Filme 14+ 2"))
quatorze_anos.add_child(Node("Filme 14+ 3"))

dezesseis_anos.add_child(Node("Filme 16+ 1"))
dezesseis_anos.add_child(Node("Filme 16+ 2"))
dezesseis_anos.add_child(Node("Filme 16+ 3"))

dezoito_anos.add_child(Node("Filme 18+ 1"))
dezoito_anos.add_child(Node("Filme 18+ 2"))
dezoito_anos.add_child(Node("Filme 18+ 3"))


def recomendar_filmes(raiz, idade_usuario):
    recomendacoes = []
    # Itera sobre as faixas etárias filhas do nó raiz. No caso [10+, 12+, 14+ 16+ 18+]
    for faixa in raiz.get_children():

        # Mostra a faixa etária atual que está sendo processada
        print(f"Processando faixa etária: {faixa.value}")
        
        # Obtém a idade mínima para a faixa etária atual
        idade_minima = obter_idade_minima(faixa.value)
        print(f"Idade mínima para {faixa.value}: {idade_minima}")
        
        #  Se a idade for minima eu paro o loop e trago apenas os filmes que ele se enquadra, evita processamento
        if idade_usuario <= idade_minima:
            break
        # Verifica se a idade do usuário é suficiente para a faixa etária
        elif idade_usuario >= idade_minima:
            # Itera sobre os filmes da faixa etária e adiciona à lista de recomendações
            for filme in faixa.get_children():
                print(f"Adicionando filme: {filme.value}")
                recomendacoes.append(filme.value)

    # Mostra a lista final de recomendações
    print(f" \n Recomendações finais: {recomendacoes}")
    return recomendacoes

 

# Faz um de-para da string para Numerico, para utilizar num IF logo acima
def obter_idade_minima(faixa_etaria): 
    idade_minima = 0
    if "Livre" in faixa_etaria:
        idade_minima = 0
    elif "10+" in faixa_etaria:
        idade_minima = 10
    elif "12+" in faixa_etaria:
        idade_minima = 12
    elif "14+" in faixa_etaria:
        idade_minima = 14
    elif "16+" in faixa_etaria:
        idade_minima = 16
    elif "18+" in faixa_etaria:
        idade_minima = 18
    return idade_minima 


# Testamos a aplicação chamando a função recomendar_filmes com diferentes idades de usuários para verificar se as recomendações estão corretas.
print(recomendar_filmes(root, 8))  # Deverá retornar filmes até a faixa de 12+
```
